## 3. Parse, merge and tidy your data so that you have a row for each state and two columns: state abbrevation, location count.

````python
# Selecting the relevant columns: 'State Abbreviation' and 'Location Count'
selected = data[['State Abbreviation', 'Location Count']]

# Grouping by 'State Abbreviation' and summing the 'Location Count' in case there are duplicates
grouped = selected.groupby('State Abbreviation').sum().reset_index()

grouped.head()
````

## 4. Scrape the state names and populations from this wikipedia page. Convert the state names to abbreviations and merge these data with your coffee dataset.

````python
# I downloaded the HTML and built the code off of it

html_file_path = '/content/view-source_https___simple.wikipedia.org_wiki_List_of_U.S._states_by_population.html'

# Read the HTML file
with open(html_file_path, 'r', encoding='utf-8') as file:
    html_content = file.read()

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find the table containing the state names and populations
table = soup.find('table', {'class': 'wikitable'})

# Extract the rows from the table
rows = table.find_all('tr')

# Initialize lists to store state names and populations
state_names = []
populations = []

# Iterate over the rows to extract state names and populations
for row in rows[1:]:  # Skip the header row
    cols = row.find_all('td')
    state_name = cols[2].get_text(strip=True)
    population = cols[3].get_text(strip=True).replace(',', '')  # Remove commas

    state_names.append(state_name)
    populations.append(int(population))  # Convert population to integer

# Display the first few state names and populations
state_names[:5], populations[:5]
````

````python
# chat gpt said i had to modify the previous function for it to work below
def stateabb(state_name, data):
    # Create a dictionary mapping state names to their abbreviations
    state_to_abbreviation = dict(zip(data['State'], data['State Abbreviation']))

    # Return the state abbreviation for the given state name
    return state_to_abbreviation.get(state_name, "Unknown")
````

````python
# Convert the scraped state names to abbreviations
state_abbreviations = [stateabb(state, data) for state in state_names]

# Create a new dataframe with the scraped data
df_scraped = pd.DataFrame({
    'State Abbreviation': state_abbreviations,
    'Population': populations
})

# Merge the scraped data with the coffee dataset
df_merged = pd.merge(grouped, df_scraped, on='State Abbreviation', how='inner')

df_merged.head()
````

## 6. Create a region variable in your dataset according to the scheme on this wikipedia page: Northeast, Midwest, South, West. You do not need to scrape this information.

````python
# Define the states in each region
northeast_states = ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 'Rhode Island', 'Vermont',
                    'New Jersey', 'New York', 'Pennsylvania']
midwest_states = ['Illinois', 'Indiana', 'Iowa', 'Kansas', 'Michigan', 'Minnesota', 'Missouri', 'Nebraska',
                  'North Dakota', 'Ohio', 'South Dakota', 'Wisconsin']
south_states = ['Delaware', 'Florida', 'Georgia', 'Maryland', 'North Carolina', 'South Carolina', 'Virginia',
                'District of Columbia', 'West Virginia', 'Alabama', 'Kentucky', 'Mississippi', 'Tennessee',
                'Arkansas', 'Louisiana', 'Oklahoma', 'Texas']
west_states = ['Arizona', 'Colorado', 'Idaho', 'Montana', 'New Mexico', 'Utah', 'Nevada', 'Wyoming',
               'Alaska', 'California', 'Hawaii', 'Oregon', 'Washington']

# Create a mapping of state abbreviations to regions
state_to_region = {}
for state in northeast_states:
    state_to_region[stateabb(state, data)] = 'Northeast'
for state in midwest_states:
    state_to_region[stateabb(state, data)] = 'Midwest'
for state in south_states:
    state_to_region[stateabb(state, data)] = 'South'
for state in west_states:
    state_to_region[stateabb(state, data)] = 'West'

# Add a Region column to the dataset
df_merged['Region'] = df_merged['State Abbreviation'].map(state_to_region)

df_merged.head()
````

## Automate

````python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_menuism(url):
    # Send a request to the URL
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract company name (example, may need to be adapted)
    company_name = url.split('/')[-1]  # Assuming the company name is the last part of the URL

    # Scrape the data (example, may need to be adapted)
    # Assume the data is in a table with specific class or structure
    table = soup.find('table', {'class': 'specific-class'})
    rows = table.find_all('tr')

    state_names = []
    location_counts = []
    for row in rows[1:]:  # Skip the header row
        cols = row.find_all('td')
        state_name = cols[0].get_text(strip=True)
        location_count = int(cols[1].get_text(strip=True))

        state_names.append(state_name)
        location_counts.append(location_count)

    # Convert state names to abbreviations
    state_abbreviations = [stateabb(state) for state in state_names]  # Use the stateabb function defined earlier

    # Organize data into a DataFrame
    data = {
        'State Abbreviation': state_abbreviations,
        'Location Count': location_counts,
        'Company Name': [company_name] * len(state_names)
    }
    df = pd.DataFrame(data)

    return df

# Example usage
# url = 'https://www.menuism.com/company-name'
# df = scrape_menuism(url)
````
