## 3. Which major geographical region sold the most total organic, small Hass avocados in 2017?

In order to accomplish this, we first need to create a new column type that allows us to differentiate the geographies.

```python
avocado_clean = data

# Classifying the geography

states = ['California', 'South Carolina']
regions = ['Midsouth', 'Northeast', 'Plains', 'South Central', 'Southeast', 'West', 'Great Lakes']

def classify_geography(geo): # ChatGPT assisted with template to incorportae if or else for multiple options.
    if geo in states:
        return "state"
    elif geo in regions:
        return "region"
    elif geo == "Total U.S.":
        return "Total U.S."
    else:
        return "metro area"

avocado_clean['geography type'] = avocado_clean['geography'].apply(classify_geography)

avocado_clean
```

```python
# Filtering data to meet requirements.
avocado_clean = avocado_clean[(avocado_clean['year'] == 2017) & (avocado_clean['type'] == 'organic') & (avocado_clean['geography type'] == 'region')]

avocado_clean
```

```python
# Now that the data is cleaned and filitered as needed, we only have the data that fits the requirements, and can now sort the data

avocado_clean[avocado_clean['geography'].isin(regions)].groupby('geography')['4046'].sum().sort_values(ascending=False)
```

The West is the major geographical area that sold the most small Hass avocados in 2017.

## 4. Split the date variable into month, day, and year variables. In which month is the highest average volume of avocado sales?

```python
# First create the new columns so that we can sort the data

data['date'] = pd.to_datetime(data['date'])

data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day
data['year_column'] = data['date'].dt.year

avocado_clean = data
avocado_clean
```

```python
avocado_clean.groupby('month')['total_volume'].mean().sort_values(ascending=False)
```

The month with the highest average volume of avocado sales was the fifth month, May.

## 5. Which metro area geographical regions sold the most total avocados? Plot side-by-side box-plots of the total volume for only the five metro geographical regions with the highest averages for the total_volume variable.

```python
# Filtering by metro area and sorting
avocado_clean5 = avocado_clean[avocado_clean['geography type'] == 'metro area']
avocado_clean5.groupby('geography')['total_volume'].sum().sort_values(ascending=False)
```

Los Angeles, New York, Dallas/Ft. Worth, Houston, & Phoenix/Tuscon regions sold the most total avocados.

```python
# Swap out total for mean
avocado_clean5.groupby('geography')['total_volume'].mean().sort_values(ascending=False)
# We obtain the same group of geogrraphies, makes sense if there is an equal amount of occurences.
```

```python
# Filter data to only include top 5.
avocado_clean5 = avocado_clean5[avocado_clean5['geography'].isin(['Los Angeles', 'New York', 'Dallas/Ft. Worth', 'Houston', 'Phoenix/Tucson'])]
avocado_clean5
```

```python
# Boxplot

plot = (
    ggplot(avocado_clean5, aes(x = 'geography', y = 'total_volume', fill = 'geography'))
    + geom_boxplot()
    + labs(title = "Top 5 Metro Area Geographical Regions(Based on Average Volume)", x="Metro Area", y="Total Volume")
    + scale_y_continuous(breaks = range(0, 6000000, 500000))
    + theme_light()
    + theme(figure_size = (10, 5))
)

plot
```

## 8. The following plot shows, for all four California regions, the proportion of the average Hass avocado sales that are small, large, or extra large; conventional vs. organic. Recreate the plot; you do not have to replicate the exact finishing touches - e.g., color, theme - but your plot should resemble the content of this plot.

```python
# Data filtered to California regions only
avocado_cleanCA
```

```python
transform_data = pd.melt(avocado_cleanCA, id_vars=['geography', 'type'],
                       value_vars=['4046', '4225', '4770'],
                       var_name='size', value_name='proportion')
```

```python
plot = (ggplot(transform_data, aes(x='geography', y='proportion', fill='size'))
        + geom_bar(aes(fill='size', group='type'), position=position_fill(), stat='identity', width=0.9)
        + facet_wrap('~ type', ncol=2)
        + labs(title='Proportion of Averages Hass Avocado Sales by Size', x='Region of California', y='Proportion', fill='Size')
        + scale_y_continuous(breaks=(0, 0.25, 0.50, 0.75, 1), labels=lambda y: ['0%', '25%', '50%', '75%', '100%'])
        + scale_fill_manual(breaks=('4770', '4225', '4046'), values=('slateblue', 'seagreen', 'chocolate'), labels = {'4770': 'XLarge', '4225': 'Large', '4046': 'Small'})
        + theme_light()
        + theme(axis_text_x=element_text(rotation=45, hjust=1)) # Chat GPT assisted with spacing/rotating region names.

)

plot
```

# Using Outside Data
## A joke in the media is that Millennials can’t afford houses, because they spend all their money buying avocado toast. Let’s use this data set to address that claim.

```python
# Loading second data set
prices = pd.read_csv('/content/variable_full_hypothetical_california_house_prices.csv')
prices
```

```python
merged_data = pd.merge(avocado_cleanCA,
                       prices[['City', 'Date', 'Average House Price ($)']],
                       left_on=['geography', 'date'],
                       right_on=['City', 'Date'],
                       how='inner')

merged_data.drop(columns=['City', 'Date'], inplace=True)

merged_data

```

```python
# Plot of House Prices vs. Avocado Volume
plot = (ggplot(merged_data, aes(x='total_volume', y='Average House Price ($)', color='geography'))
        + geom_point(alpha=0.6)
        + labs(title="House Prices vs. Avocado Volume", x="Total Volume of Avocados Purchased", y="Average House Price ($)")
        + theme_light()
        + scale_color_discrete(name="California Region")
        + theme(figure_size=(10, 5))
       )

plot
```

As seen above, there does not appear to be a relationship between the amount of avocados purchased, and the average house price in either region. Whether a region purhcases a lot or a little avocados, it does not appear to have a significant change or trend in the average house price within the region. The same appears to hold true for prices below.

```python
# Plot of House Prices vs. Avocado Prices
plot = (ggplot(merged_data, aes(x='average_price', y='Average House Price ($)', color='geography'))
        + geom_point(alpha=0.6)
        + labs(title="Average House Prices vs. Average Avocado Prices", x="Average Price of Avocados Purchased", y="Average House Price ($)")
        + theme_light()
        + scale_color_discrete(name="California Region")
        + theme(figure_size=(10, 5))
       )

plot
```
