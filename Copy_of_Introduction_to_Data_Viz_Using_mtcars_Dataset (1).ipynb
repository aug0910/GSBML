Use the `astype()` method to convert the categorical variables of the `mtcars` (df) dataset to have text values in the code chunk below. We've done one for you!

It should also be clear by how `plotnine` treats these variables.

1. Create side-by-side boxplots of the `mpg` variable by the different values of the `am` variable. What happens if you convert the `am` variable back to float values and then try to create this same plot? Explain the differences between the two plots.

2. Create overlaid histograms of the `hp` variable for the different values of the `cyl` variable. What happens if you convert the `cyl` variable back to float values and then try to create this same plot? Explain the differences between the two plots.

```python
# Create plot for (1) here
from plotnine import ggplot, geom_point, aes, geom_boxplot, geom_boxplot, labs, geom_histogram, theme_light, scale_fill_manual


# Create the boxplot
plot = (
    ggplot(mtcars, aes(x='factor(am)', y='mpg', fill = "am")) +
    geom_boxplot() +
    labs(x='Transmission (0=Automatic, 1=Manual)', y='Miles Per Gallon (mpg)', title='MPG by Transmission Type')
)

plot

# If am is swithced back to float values, the x axis will be adjusted to show the values as continous rather than discrete. For example, if you put a # in front where it would normally be a categorical variable, the colors are not seperate. Instead it turns into a gradient which probably isn't the best option.
```

```python
# Plot (2) here

plot_int_cyl = (
    ggplot(mtcars, aes(x='hp', fill='factor(cyl)')) +
    geom_histogram(position="identity", alpha=0.5, bins=30) +
    labs(x='Horsepower (hp)', y='Count', fill='Cylinders', title='Overlaid Histograms of hp by cyl') +
    theme_light() +
    scale_fill_manual(values=["red", "green", "blue"])
)

plot_int_cyl

# If cyl is swithced back to float values, the x axis will be adjusted to show the values as continous rather than discrete. For example, if you put a # in front where it would normally be a categorical variable, the colors are not seperate. Instead it turns into a gradient which probably isn't the best option since there are only a few cylinder options

```

Some of these variables, like the `cyl` variable, have numeric values that actually make sense as numbers (i.e. the number of cylinders in the engine). However, it doesn't make the most sense to "do math" with this type of variable (e.g. take averages and such) because there are so few different value this can take on AND they're an explicit choice made by the car manufacturer. So, it makes more sense to treat the `cyl` variable as a categorical variable despite it having numeric values.

You will need to keep these kinds of nuances about data in mind as you work with an increasing variety and richness of data, and do more complex things with them.

Choose 3 new `geometries` from the [data-to-viz website](https://www.data-to-viz.com/) for graphs that you'd like to explore using the `mtcars` dataset, and then create those graphs!

```python
# Create plot with first new geometry here
# Converting #2 into a scatterplot, could be a better visual instead of overlaid histogram since you can see the trends

from plotnine import scale_color_manual

plot_scatter = (
    ggplot(mtcars, aes(x='hp', y='mpg', color='factor(cyl)')) +
    geom_point() +
    labs(x='Horsepower (hp)', y='Miles per Gallon (mpg)', color='Cylinders', title='Scatterplot of mpg vs. hp colored by cyl') +
    theme_light()
)

plot_scatter
```

```python
# Create plot with second new geometry here
# Turning #1 into a density plot to show another visualization that may better represent each transmission type. Chat gpt assisted with the geom_density for transparency so that you can see both graphs even when overlaid
from plotnine import geom_density

plot_density = (
    ggplot(mtcars, aes(x='mpg', fill='factor(am)')) +
    geom_density(alpha=0.5) +
    labs(x='Miles Per Gallon (mpg)', y='Density', fill='Transmission (0=Automatic, 1=Manual)', title='Density Plot of MPG by Transmission Type') +
    theme_light() +
    scale_fill_manual(values=["red", "blue"])
)

plot_density

```
